```python
# 导入必要的库
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import time
import warnings
import copy
import os
warnings.filterwarnings("ignore")

# 从 xsmom.py 文件中导入 XSMOM 类
from xsmom import XSMOM
```

```python
# 假设这些数据已经被加载到内存中
# trade_dates, zz500, total_return, total_price, target_stock 需要根据实际数据替换
zz500 = pd.read_csv('zz500.csv', index_col=0)
zz500.index = pd.to_datetime(zz500.index)
trade_dates = zz500.index
total_return = pd.read_pickle('total_return.pkl')
```

```python
# 初始化回测框架
xsmom = XSMOM(total_return, zz500, train_window=20)
```

```python
# 运行回测
xsmom.run(
    get_Xt=np.max,
    start_day=20200110,
    end_day=20231229,
    skip_first_day=True,
    step=5,
    long_short='all',
    note=None
)
```

```python
# 显示策略表现
strategy_return, strategy_signal = xsmom.cal_strategy(
    period=len(trade_dates),
    step=5,
    start_day_index=0,
    skip_first_day=True,
    get_Xt=np.max
)
bench_mark = zz500['收盘'].iloc[::5]
xsmom.show_strategy(strategy_return, strategy_signal, bench_mark, long_short='all', step=5, note='example_strategy')
```
